# Solving 1D Poisson equation with finite elements

We are given the 1D Poisson equation as 

\begin{equation}  
    -u_{xx} = f(x),
\end{equation}

in the domain $\Omega: x \in [0,1]$. At the boundaries we have Neumann conditions at $x=0$ given as $u_x = C$ where C is a nonzero constant, and Dirichlet condition at $x=1$ given as $u = D$, where D is a nonzero constant. 

## Task 1


Here we will find the weak variational form of the PDE. First, because we have a nonzero Derichlet condition, we can have a boundary function $B(x) = Dx$ to our proposed solution. Thus we have that 

\begin{equation}
u(x) = B(x) + \sum_{j \in I_s} c_j\psi_j(x),
\end{equation}
where $I_s$ is the index set. 

To find the weak variational form, we use the Galerkin method, to minimize the error. Thus, we want to find 

\begin{equation}
(R, \psi_i)=0, \forall i \in I_s.
\end{equation}

The residue R is given as $u_{xx} - f(x) = 0$, and so we get 

\begin{equation}
\begin{aligned}
(u_{xx} + f, \psi_i) &= (u_{xx}, \psi_i) + (f, \psi_i)\\
                     &= -(u_x, \psi'_i) + [u_x\psi_i]_0^1 + (f, \psi_i)\\
                     &=  -(u_x, \psi'_i) + u_x(1)\psi_i - u_x(0)\psi_i + (f, \psi_i) \\
                     &= -(u_x, \psi'_i)- C\psi_i + (f, \psi_i). 
\end{aligned}
\end{equation}

Thus we need to find the $\psi_i, \forall i \in I_s$ such that we get 

\begin{equation}
(u_x, \psi'_i) =  (f, \psi_i) - C\psi_i, \forall i \in I_s.
\end{equation}

Insert that 
\begin{equation}
    u(x) = B(x) + \sum_j c_j \psi_j = D\psi_{N_e} + \sum_j^{N_e-1}c_j\psi_j
\end{equation}

which gives the updated weak variational form

\begin{equation}
(D\psi_{N_e} + \sum_j^{N_e-1}c_j\psi_j, \psi'_i) =  (f, \psi_i) - C\psi_i
\end{equation}

\begin{equation}
\sum_j^{N_e-1}c_j(\psi_j, \psi'_i) =  (f, \psi_i) - C\psi_i - (D\psi_{N_e}', \psi_i')
\end{equation}

which gives us 

\begin{equation}
    A_{i,j} = (\psi_j, \psi'_i)
\end{equation}

and 

\begin{equation}
    b_{i} = (f, \psi_i) - C\psi_i - (D\psi_{N_e}', \psi_i')
\end{equation}

## Task 2

In the case of the first element we have three nodes, since we use $P^2$ function space. Our standardized Lagrange functions are thus given as 

\begin{equation}
\tilde{\phi}_0 = \frac{1}{2}(X-1)X
\end{equation}
\begin{equation}
\tilde{\phi}_1 = (1-X^2)
\end{equation}
\begin{equation}
\tilde{\phi}_2 = \frac{1}{2}(X+1)X.
\end{equation}

These relations come from the affine mapping from x to X, given as 

\begin{equation}
x = x_m + \frac{1}{2}X,
\end{equation}

where $x_m = (x_r + x_l)/2$ and $h = x_r - x_l$. 

The matrix contribution will here be a three by three matrix. The elements $A_{i,j}$ are given as 

\begin{equation}
A_{i,j} = \int_{\Omega} \phi_{i,x}\phi_{j,x} dx.
\end{equation}

Instead of integrating over the whole domain, where the basis functions for the most part are zero, we can instead use the transformation above, solving elementwise matrices and assembling them afterwards. For a given element we thus have $\tilde{A}^{(e)}$ given as 


\begin{equation}
\tilde{A}_{r,s}^{(e)} = \int_{-1}^1 \phi_{q(e,r),x}\phi_{q(e,s),x} \frac{h}{2} dX.
\end{equation}

We have to remember the fact that we wish to derive the basis functions with respect to X, and not x. Such a derivation for a given function is given as 

\begin{equation}
    \frac{d\tilde{\phi}_r}{dx} = \frac{d\tilde{\phi}_r}{dX} \frac{dX}{dx} = \frac{2}{h} \frac{d\tilde{\phi}_r}{dX},
\end{equation}

which leaves with 
\begin{equation}
\tilde{A}_{r,s}^{(e)} = \frac{2}{h}\frac{2}{h}\int_{-1}^1 \frac{d\phi_{q(e,r)}}{dX}\frac{d\phi_{q(e,s)}}{dX} \frac{h}{2} dX.
\end{equation}

Thus, we solve these integrals and get 

\begin{equation}
\tilde{A}^{(0)} = 
\frac{1}{3h}
\begin{bmatrix}
7 & -8 & 1\\
-8 & 16 & -8 \\
1 & -8 & 7
\end{bmatrix},
\end{equation}



with the Neumann condition we get that the first vector element in b vector have an additional C subtracted in. This is shown from the weak variational form.

## Task 3

As with before we use the standardized basis functions, but here we use $P^1$ basis functions. They are given as 

\begin{equation}
P^0 = \frac{1}{2}(1-X)
\end{equation}

\begin{equation}
P^1 = \frac{1}{2}(1+X).
\end{equation}

Solving the matrix elements for the last sub element of the domain thus gives us

\begin{equation}
\tilde{A}^{(-1)} = 
\frac{1}{h}
\begin{bmatrix}
1 
\end{bmatrix}.
\end{equation}
This is because we have one value that is known, i.e the the Dirichlet boundary condition. Thus we only get an unknown contribution from the left node in the last element. 

Here we have to remember that the Dirichlet boundary condition will add another value D to the last vector element in vector b. This is shown from the weak variational form. 

## Task 4

Here we will find a analytical solution to the 1D Poisson equation, for a given $f(x)$, C and D. Here I choose the function 

\begin{equation}
f(x) = 2.
\end{equation}

We thus have that 

\begin{equation}
    -u_{xx} = 2
\end{equation}

We begin by integrating both sides twice, and get

\begin{equation}
u(x) = c_2 x + c_1 -x^2
\end{equation}



Now we insert the boundary conditions, namely that $u_x(0) = C$ and $ u(1) = D$. We thus get that $c_2 = C$ and $c_1 = D-C+1$. Thus we have the exact solution given as 

\begin{equation}
u(x) = Cx + D-C+1 - x^2
\end{equation}


## Task 5

### Import libraries

```python
import matplotlib.pyplot as plt 
import numpy as np
```

### Declare functions

```python
#Basis functions in standardized form
def phi_0_quad(X):
    return 0.5*(X-1)*X
def phi_1_quad(X):
    return 1-X**2
def phi_2_quad(X):
    return 0.5*(X+1)*X
def phi_0_lin(X):
    return 0.5*(1-X)
def phi_1_lin(X):
    return 0.5*(1+X)

#Derivatives of standardized basis functions
def phi_0_quad_der(X):
    return (X-0.5)
def phi_1_quad_der(X):
    return -2*X
def phi_2_quad_der(X):
    return X+0.5
def phi_0_lin_der(X):
    return -0.5
def phi_1_lin_der(X):
    return 0.5


#External functions and information

def x_val(e,X):
    return xm[e] + h/2*X

def f(x):
    return 2  

def exact(x):
    return C*x + D + 1 - C - x**2


def integrate(f1, f2, N, e):
    X = np.linspace(-1, 1, N+1)
    if f1 == "f":
        f1x = f1(x_val(e, X))
    elif f2 == "f":
        f2x = f2(x_val(e, X))
    else:
        f1x = f1(X)
        f2x = f2(X)
    area = np.sum(f1x*f2x)*h/N
    return area
```

### Create A matrix

```python
elements = 20

place = 1/elements

element_placement = []
start = 0
for i in range(1, elements+1):
    element_placement.append( [start, i*place] )
    start = i*place
    
#print(element_placement)
    
xm = np.zeros(len(element_placement))
x_arr = [0]

for i, elem in enumerate(element_placement):
    xm[i] = 0.5*(elem[0] + elem[1])
    if i < int(elements/2):
        x_arr.append(xm[i])
        x_arr.append(elem[1])
    else:
        x_arr.append(elem[1])
    
x_arr = np.array(x_arr)



xm = np.asarray(xm)
h = np.max(x_arr)/elements

nodes = int(3/2*elements + 1)
D = 1.4
C = 0.3  
A = np.zeros((nodes-1, nodes-1))


Atilde_0 = np.array([[7, -8, 1], [-8, 16, -8], [1, -8, 7]])
Atilde_1 = np.array([[3, -3], [-3, 3]])


#Create A matrix
for e in range(0, elements, 2):
    for i in range(3):
        for j in range(3):
            A[e+i, e+j] += Atilde_0[i, j]
            
for e in range(int(elements), nodes-2):
    for i in range(2):
        for j in range(2):
            A[e + i, e + j] += Atilde_1[i, j]
            
A[-1,-1] += 3
A = (1/(3*h))*A
A_inv = np.linalg.inv(A)
```

### Create b vector

```python
quad_der = [phi_0_quad_der, phi_1_quad_der, phi_2_quad_der]
quad = [phi_0_quad, phi_1_quad, phi_2_quad]
lin_der = [phi_0_lin_der, phi_1_lin_der]
lin = [phi_0_lin, phi_1_lin]

b = np.zeros(nodes-1) #holds the b vector elements, will be assembled


mid = int(elements/2)
dof_map = []
last_i = 2
for e in range(1, elements+1):
    dof = []
    if e == 1:
        for i in range(3):
            dof.append(i)
    elif (e <= mid) and (e != 1):
        for i in range(3):
            dof.append(last_i + i)
    elif (e >= mid) and (e != elements):
        for i in range(2):
            dof.append(last_i + i)
    else:
        dof.append(last_i)
    last_i = dof[-1]
    dof_map.append(dof)
    
print(dof_map)

for elem in range(int(elements/2)):
    b_index = dof_map[elem]
    for i, ind in enumerate(b_index): 
        b[ind] += integrate(f, quad[i] , 10001, elem)
        

for elem in range(int(elements/2), elements):
    b_index = dof_map[elem]
    for i, ind in enumerate(b_index): 
        b[ind] += integrate(f, lin[i] , 10001, elem)
        
diri_cond = C 
b[0] -= diri_cond

b[-1] -= D*2/h*(-1/2) #Integrate D phi´_N * phi´_i = 1/jacobi_det *(-1/2)
print(b)
```

### Find constants

```python
constants = A_inv@b
constants = np.append(constants, D)
print(constants)
```

### Test coefficiants

```python
#plt.plot(x_arr, constants)
```

### Show the final solution

```python
solution = []
X = np.linspace(-1, 1, 101)
for elem in range(int(elements/2)):
    b_index = dof_map[elem]
    sol = np.zeros(len(X))
    for i, ind in enumerate(b_index): 
        sol += constants[ind]*quad[i](X)
    solution.append(sol)  


for elem in range(int(elements/2), elements):
    b_index = dof_map[elem]
    sol = np.zeros(len(X))
    for i, ind in enumerate(b_index):
        sol += constants[ind]*lin[i](X)
    if elem == (elements-1):
        sol += constants[-1]*lin[-1](X)
    solution.append(sol)
    


solution = np.concatenate(solution)


x_data = np.linspace(0,1, len(solution))

fig = plt.figure(figsize=(14,14))
plt.plot(x_data, exact(x_data), label="Exact")
plt.plot(x_data, solution, label="Approx")
plt.legend()
plt.show()
```

```python
er = np.abs((exact(x_data) - solution))
fig = plt.figure(figsize=(14,14))
plt.plot(x_data, er)
plt.show()
```
